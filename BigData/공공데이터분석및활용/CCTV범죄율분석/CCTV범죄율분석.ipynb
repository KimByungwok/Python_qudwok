```python
# 라이브러리 로드
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### CCTV 데이터 읽기

```python

# CCTV 데이터 읽기
#CCTV = pd.read_excel('서울시CCTV설치운영현황(자치구)_년도별_211231기준.xlsx')
CCTV = pd.read_excel('seoul_cctv.xlsx')
```

```python
# 읽은 데이터 확인
# 확인된 사항에 대해 이야기해 보기
CCTV
```

```python
# 데이터 출력으로 확인하기 어려운 경우
# 데이터프레임 구조 및 기술통계 확인
CCTV.info()
CCTV.describe()
```

```python
# 확인된 사항을 바탕으로 데이터 수정
# CCTV의 결측 데이터 메꾸는 작업 수행
# 필드 성격에 따라 어떤값을 써야하는 지 생각해보기
# 여기서는 NaN --> 0 으로
CCTV.fillna(0, inplace=True)
```

```python
# 데이터 수정 확인
CCTV
```

```python
# 컬럼 이름 '기관명' --> '구청별'로 변경
CCTV.rename(columns={CCTV.columns[0]: '구청별'}, inplace=True)
CCTV.head(2)
```

```python
# 컬럼 정리
# - 2016년 이전 자료를 모두 합쳐 '2016년 이전'이라는 컬럼 이름으로 저장
# - 정리 후 합치기 전의 2016년 이전 컬럼은 삭제
CCTV['2016년 이전'] = CCTV['2011년 이전'] + CCTV['2012년'] + CCTV['2013년'] +\
CCTV['2014년'] +CCTV['2015년'] + CCTV['2016년']
CCTV = CCTV.drop(['2011년 이전', '2012년', '2013년', '2014년', '2015년', '2016년'], axis=1)
CCTV.head(2)
```

```python
# 소계 컬럼을 기준으로 오름차순 정렬해보기
CCTV_sorted_asc = CCTV.sort_values(by='소계', ascending=True)
CCTV_sorted_asc
```

```python
# 소계 컬럼을 기준으로 내림차순 정렬해보기
CCTV_sorted_desc = CCTV.sort_values(by='소계', ascending=False)
CCTV_sorted_desc
```

### 서울시 거주 인구 데이터 읽기

```python
# 서울시 거주 데이터 파일 읽고 데이터 프레임 확인
POPU = pd.read_excel('seoul_popu.xlsx')
POPU.head()
```

```python
# 자치구, 인구총합, 외국인 총합, 65세 고령자 총합 컬럼만으로 구성하도록 읽기
# read_excel()의 usecols 인수 활용, 무슨값을 줘야 하는지 생각해 보기
POPU = pd.read_excel('seoul_popu.xlsx', header=2, usecols='B, D, G, J, N')
POPU
```

```python
# 일단 전체 읽은 후 자치구, 인구총합, 외국인 총합, 65세 고령자 총합 컬럼만으로 구성하도록 정리
# 데이터 프레임의 iloc() 사용해 슬라이싱, 무슨값을 줘야 하는지 생각해 보기
POPU = pd.read_excel('seoul_popu.xlsx')
POPU = POPU.iloc[2:, [1, 3, 6, 9, 13]]
POPU
```

```python
# 열 라벨 이름을 순서대로 '구청별', '총인구', '한국인', '외국인', '고령자'로 변경
#POPU.columns = pd.Index(['구청별', '총인구', '한국인', '외국인', '고령자'])
#POPU.head(2)
```

```python
POPU = POPU.rename(columns={POPU.columns[0]: '구청별', POPU.columns[1]: '총인구',
                            POPU.columns[2]: '한국인', POPU.columns[3]: '외국인',
                            POPU.columns[4]: '고령자'})
POPU.head()
```

```python
# CCTV와 POPU를 합치기 위해 POPU의 데이터를 조정
# 구청별 컬럼의 값 중 "합계"는 구청명이 아님
POPU.drop(2, inplace=True)
POPU
```

```python
# 인덱스 범위 조정
# 행 삭제 이후 나머지 행들에 대해 인덱스를 새로 부여

POPU.index = pd.Index(np.arange(0, 25, 1))
POPU
```

```python
# 서울시에서 외국인 많은 곳?
POPU.sort_values(by='외국인', ascending=False).head(3)
```

```python
# 서울시에서 고령자 많은 곳?
POPU.sort_values(by='고령자', ascending=False).head(3)
```

### 데이터 합치기

```python
# CCTV와 POP를 합쳐보자
# 결과 확인 후 이상한 점 찾기
pd.merge(POPU, CCTV, on='구청별')
```

```python
# 구청별 데이터의 동질성(homogeneity) 위배 해결
for df in (POPU, CCTV):
    df['구청별'] = df['구청별'].str.replace('\s+', '')
```

```python
# 다시 CCTV와 POP를 합쳐보자
merged = pd.merge(POPU, CCTV, on='구청별')
merged
```

```python
# 구청별로 세부 내용을 처리할 것이므로 '구청별' 열을 인덱스로 지정
merged.set_index('구청별', inplace=True)
merged
```

```python
# 1인당 CCTV설치 비율 계산해 '1인당CCTV설치비율'열 추가
merged['1인당CCTV비율'] = merged['소계']/merged['총인구'] * 100
merged
```

```python
# 데이터간의 상관관계 분석해 보기
# ㅇ corr() 함수 사용
#   - pearson
#     : 선형상관관계 분석에 사용
#     : +1(양의 선형성) ~ -1(음의 선형성), 0은 선형관계 없음
#   - spearman
#     : 단조성 판단, 두 비교 대상의 동질성 평가(같이 증가, 같이 감소 등), 1차함수
#     : 0(전혀 상관없음)~1(100% 상관있음)
#   - kendall
#     : 단조성 판단에 사용, 일치하는 쌍의 수와 일치하지않은 쌍에 대한 동질성
#     : 0(단조 상관성 전혀없음) ~ 1(단조 상관성 100%)


# CCTV 설치 대수 대비 총인구수, 외국인 수
print('CCTV 설치 대수와 총인구수 관계 :{}'
.format(merged['총인구'].corr(merged['소계'], method='spearman')))
```

```python
# CCTV 설치 대수 대비 총인구수, 외국인 수
print('CCTV 설치 대수 대비 총인구수 : {}'.format(
merged['외국인'].corr(merged['소계'], method='spearman')))
```

### 구별 CCTV 설치 대비 범죄율 상관관계 분석

```python
# 서울시 5대범죄 발생현황 통계 데이터 읽기
# CSV 파일에 데이터 구분자로 tab이 사용되었을 때 처리
CRIME = pd.read_csv('seoul_crime.csv', sep='\t')
CRIME
```

```python
# 분석에 필요한 데이터 삭제
# 자치구, 합계, 합계.1 컬럼을 남기고 나머지 컬럼 삭제
# 1,2행은  필요없는 부분이므로 삭제
CRIME = CRIME.iloc[2:, 1:4]
CRIME.head()
```

```python
# 열라벨 이름 변경
CRIME.columns = pd.Index(['구청별', '범죄건', '검거건'])

# 구청별 열라벨을 행의 인덱스로 지정
CRIME.set_index('구청별', inplace=True)
CRIME.head(3)
```

```python
# 범죄 검거율을 열에 추가
# 데이터 셋의 요소를 조사해 수치계산이 가능한 지 확인 필요
# --> 자료형을 변경
CRIME = CRIME.apply(lambda x: x.str.replace(',','').apply(pd.to_numeric))
CRIME['검거율'] = CRIME['검거건']/CRIME['범죄건'] * 100
CRIME
```

```python
# merged와 CRIME을 결합
merged1 = pd.concat([merged, CRIME], axis=1)
merged1
```

> 서울시 구별 총 CCTV 설치 수 소계와 범죄건, 검거건의 상관관계 살펴보기
- 어떻게 해석할 것인가 생각해 보기

```python
plt.figure()
axsub = merged1.plot.scatter(x='검거건', y='소계', color='red', s=50,
                             label='검거데이터');
merged1.plot.scatter(x='범죄건', y='소계', color='blue', label='범죄데이터', s=30,
                     ax=axsub);
plt.xlabel('검거건/범죄건')
plt.show()
```
