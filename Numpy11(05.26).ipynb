```python
# 주가는 시계열 데이터(시간의 변화에 따른 값들의 집합)로 시간에 따른 변화 추세와 패턴을 분석할 필요 
# 수집한 기간동안의 데이터를 월별로 리샘플링하고 평균을 계산해 시각화
'''
DataFrame.resample(rule,                # 대상 변환을 나타내는 오프셋 문자열 또는 개체
                   axis=0,              # 업 샘플링 또는 다운 샘플링에 사용할 축
                   closed=None,         # 빈 간격의 어느 쪽이 닫혀 있는지 지정
                   label=None,          # 버킷에 레이블을 지정할 빈 가장자리 레이블을 지정
                   convention='start',  # PeriodIndex의 경우 start,end, s 또는 e 
                   kind=None,           # 결과 색인의 종류,  timestamp 또는 period
                   loffset=None,        # 리샘플링 된 시간 레이블을 조정
                   base=None,           # 정수
                   on=None,             # 리샘플링을 위해 인덱스 대신 사용할 datetime 형을 가지는 열
                   level=None,          # 리샘플링에 사용할 레벨의 이름, datetime과 유사해야 함
                   origin='start_day',  # 그룹화를 조정할 타임 스탬프
                   offset=None)         # origin매개 변수에 추가 된 오프셋으로 timedelta
'''
ss['Adj Close'].resample('M').mean().plot(kind='bar', figsize=(15, 5))
plt.show()
```

```python
# 삼성전자 주식에 대해 주가 변동률(change) 구하기
# 주가 변동률 = 전일 마지막 매매가와 현재 주가 간 비율
'''→←↑↓
DataFrame.shift(periods=1,        # 정수 값으로 axis에 따라 음수(↑ 또는 ←) 또는 양수(↓ 또는 →)일
                 freq=None,       # DateOffset,tseries.offsets,timedelta 또는str. 데이터를 재정렬하
                 axis=0,          # 이동 기준 : 행(axis=0) 또는열(axis=1)
                 fill_value=None) # 새로 도입 된 결측값에 사용할 스칼라 값
'''
# 일단 처리대상인 조정 종가(Adj Close)만 추출
ss_adj = pd.DataFrame(ss['Adj Close'])
# log() 함수를 적용해 큰 스케일을 작은 스케일로 변환
ss_adj['change'] = np.log(ss_adj['Adj Close']/ss_adj['Adj Close'].shift())
# 삼성전자 주식 주가 변동률 출력
# 잡음성 그래프로 분석이 힘듦 --> 변동성 그래프로 변경 필요
plt.figure(figsize=(10, 7))
plt.plot(ss_adj.change, lw=1)
plt.show()
```

```python
# 변동성(volatility) 그래프로 변환하기 위해 데이터 롤링
# 데이터 롤링
#   - 특정 시간 시리즈에 대한 데이터를 분석하고 해당 데이터에서 추세를 파악
# 변동성
#   - 통계에서 변동성은 유가 증권이나 시장 지수의 수익을 분산하는 통계
#   - 수익의 표준편차 또는 분산을 의미
#   - 변동성이 높을수록 안정성 저해(불안정)
'''
# 시계열 데이터에 대한 이동 평균 구하는 함수
DataFrame.rolling(window,           # 정수, 오프셋 또는 BaseIndexer 서브 클래스 유형 매개 변수로 창 
                  min_periods=None, # 정수, 윈도우의 최소 관측치 수
                  center=False,     # 불값, 창 중앙에 레이블을 설정할 지 여부 
                  win_type=None,    # 문자열, 창 유형을 지정
                  on=None,          # 문자열, 인덱스가 아닌 롤링 윈도우를 계산할 열 이름을 지정
                  axis=0,           # 정수 또는 문자열
                  closed=None)      # 문자열, 간격 마감(오른쪽, 왼쪽, 둘 다 또는 둘 다 없음)
'''
# 삼성전자 주식에 대해 변동성 그래프를 그려봄
ss_adj['volatility'] = ss_adj.change.rolling(20).std()  # 20개씩 묶어서 분산 구함
ss_adj['volatility'].plot(figsize=(10, 7), lw=1)
plt.show()
```

```python
# 일반적으로 변동성이 큰 주가의 추세나 패턴을 측정하기 위해 5일, 20일, 60일 120일 단위로 평균을 구함
# 데이터 롤링 기간을 달리해 각각의 평균을 구해봄
ra5 = ss['Adj Close'].rolling(window=5).mean()
ra20 = ss['Adj Close'].rolling(window=20).mean()
ra60 = ss['Adj Close'].rolling(window=60).mean()
ra120 = ss['Adj Close'].rolling(window=120).mean()
ra240 = ss['Adj Close'].rolling(window=240).mean()
# 데이터 롤링 기간별 평균 시각화
plt.figure(figsize=(10, 6))
plt.plot(ss.index, ss['Adj Close'], label='Adj Close', lw=1.5)
plt.plot(ss.index, ra5, label='5일선', lw=1.5)
plt.plot(ss.index, ra20, label='20일선', lw=1.5)
plt.plot(ss.index, ra60, label='60일선', lw=1.5)
plt.plot(ss.index, ra120, label='120일선', lw=1.5)
plt.plot(ss.index, ra240, label='240일선', lw=1.5)
plt.legend(loc='best')
plt.show()
```

```python
# 3개 사의 주식 데이터 합치기
ss1 = web.DataReader('005930.KS', 'yahoo', start, end)
lg1 = web.DataReader('066570.KS', 'yahoo', start, end)
sk1 = web.DataReader('000660.KS', 'yahoo', start, end)
ss1 = ss1['Adj Close']
lg1 = lg1['Adj Close']
sk1 = sk1['Adj Close']
comp = pd.concat([ss1, lg1, sk1], axis=1)
comp.columns = pd.Index(['삼성전자', 'LG전자', 'SK하이닉스'])
comp
```

```python
# 수익률 계산과 상관계수 구하기
#   - 수익률 계산은  pct_change() 함수 사용
#   - 상관계수는 corr() 함수 사용
pctcomp = comp.pct_change()
corr = pctcomp.corr()


#  수익률 출력해 보기
pctcomp
```

```python
# 상관계수 출력해 보기
corr
```

```python
# 숫자로 보니 잘 비교가 안되는 것 같음 --> 산포도로 시각화
# 삼성전자 vs SK하이닉스 수익 비교
plt.figure(figsize=(10, 7))
plt.scatter(pctcomp.삼성전자, pctcomp.SK하이닉스)
plt.xlabel('삼성전자 수익', fontsize=15)
plt.ylabel('SK하이닉스 수익', fontsize=15)
plt.grid()
plt.show()
```

```python
# LG전자 vs SK하이닉스 수익 비교
plt.figure(figsize=(10, 7))
plt.scatter(pctcomp.LG전자, pctcomp.SK하이닉스)
plt.xlabel('LG전자 수익')
plt.ylabel('SK하이닉스 수익')
plt.show()
```

```python
# 삼성전자 vs LG전자 수익 비교
plt.figure(figsize=(10, 7))
plt.scatter(pctcomp.삼성전자, pctcomp.LG전자)
plt.xlabel('삼성전자 수익')
plt.ylabel('LG전자 수익')
plt.show()
```

```python
# 대각선에 대해서는 커널밀도추정(kde; kernel density estimation) 적용
# - 시계열 데이터의 무작위적인 변화량을 줄이는 데이터 평활화(smoothing) 방법의 일종
# - 어떤  변수의 확률 분포를 평가하는 방법
# - 한정된 데이터 샘플을 기반으로 모집단에 대한 추론
scat = pd.plotting.scatter_matrix(pctcomp, diagonal='kde')
for ax in scat.ravel():
    ax.set_xlabel(ax.get_xlabel(), fontsize=14)
    ax.set_ylabel(ax.get_ylabel(), fontsize=14)
    ax.tick_params(labelsize=14)
plt.show()
# 아래 그림을 보고 해석해 보아라!
```

```python
# 양수의 상관관계를 살펴보기 위해 열지도(heat map)으로 표현해 보자
plt.imshow(corr, cmap='hot', interpolation='none')
plt.colorbar()
plt.grid(True)
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.show()
```
