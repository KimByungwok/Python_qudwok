# 파일 구성 형태별 외부 파일 읽어오기
    - CSV 파일 읽기
    : read_csv(filepath 또는 buffer, header=?, index_col=?)
    : header 옵션은 열(세로) 이름이 되는 행을 지정
    : index_col 옵션은 행 주소가 되는 열을 지정
    : 여러 가지 옵션

# 옵션
    - filepath_or_buffer : 파일의 위치, URL
    - sep(or delimiter) : 텍스트 형 데이터를 구분하기 위한 문자
    - header : 열(세로) 이름이 되는 행의 번호 (기본 0), 첫 행부터 데이터가 있는 경우 None
    - index_col : 행 주소가 되는 열(세로)의 번호 또는 이름
    - names : 열(세로) 이름으로 쓸 문자열 리스트
    - skipselected_rows : 처음 몇 줄을 skip인지 숫자 지정, skip할 행 번호 리스트
    - parse_dates : 날짜를 표현한 문자열을 datetime64로 변환할 지 여부 (기본 : false)
    - skip_footer : 마지막 몇 줄 skip할지 숫자로 지정
    - encoding : 텍스트 인코딩 종류 지정4

    - Excel 파일 읽기
    : read_excel(filepath or buffer)
    - json, html, 등등 같은 방법임ㅇㅇ

# 데이터의 특성 파악        39p
    - 데이터의 특성을 알아보기 의해 판다스에서 제공하는 메서드들을 사용하면 쉽게 데이터에 대한 통계정보를 활용할 수 있다

# 데이터의 집중 경향 알아보기
    - 평균값(mean)
    : 모든 데이터를 합한 후 순서상 가운데에 있는 수
    - 중앙값(median)
    : 전체 데이터를 정렬한 후 순서상 가운데에 있는 수
    : 짝수개의 데이터인 경우 가운데 위치한 두 수의 평균을 중앙값으로 함
    : 자료의 편차가 큰 값을 일부 포함하고 있는 경우 사용
    - 최빈값(mode)
    : 데이터 집합에서 출현 빈도가 가장 높은 값
    : 최빈값의 갯수는 0개 이상

# 데이터의 퍼진 정도(분포 형태)
    - 평균편차(mean deviation)
    : 절대 편차(absolute deviation) 또는 평균 절대편차(mean absolute deviation)또는
    average absolute deviation)라고도 함
    : 평균과 개별 값 사이 거리의 평균
    : 이상치로 인한 문제점 보완 가능
    : 모집단 평균편차, 표본 평균편차
    - 표준편차(standard deviation)
    : 자료의 산포도를 표현하는 수치
    : 분산의 양은 제곱근으로 정의
# 그림 39p 참고
    - 분산(variable)
    : 차이값 제곱의 평균
    : 모분산(population variance) o2은 모집단의 분산 : 관측된 값에서 모 평균을 빼서 제곱한 값을 모두 합해 전체 데이터 갯수 n으로 나눈 값
    - 상관계수(correlation coefficient)
    : 두 변수간의 통계적인 관계 표현을 위해 특정한 상관관계 정도를 수치 값으로 나타낸 계수
    : -1 ~ +1 사이의 값
