```python
import pandas as pd
```

```python
# 공에 대한 정보를 저장하고 있는 엑셀파일 두 개를 읽어
# 각각 df1, df2로 저장하라

df1 = pd.read_excel("sample/ball_data_01.xlsx")
df2 = pd.read_excel("sample/ball_data_02.xlsx")
```

```python
# 생성한 df1과 df2를 하나의 데이터 프레임(df)으로 합친 후 출력하라

df = pd.concat([df1,df2])

df
```

```python
# 구성한 데이터 프레임 df의 자료를 보고 누락자료를 처리하라
# 누락자료를 확인하기 위한 코드

df.info()
```

```python
# 확인한 누락자료의 처리 코드
# 누락자료 처리를 위해 선택한 방법에 대해 타당한 이유를 설명하라

# 컬러에 문자열이 있기 때문에 누락 자료를 처리하는 방법 중 fillna(0)으로 NaN값을 0으로 주기보다는 dropna()로 NaN 값이 있는 행을 날려버리는게 더 현명하다고 판단함.

df = df.dropna()
df.info()
```

```python
# 데이터 프레임(df)에서 색(color)의 종류는 몇가지가 있는 지 알아보아라.
# 출력은 결과 화면과 같이 나오도록 한다.

df_c = df['color'].drop_duplicates()
print("색 종류는 4가지, "+df_c.iloc[0]+","+df_c.iloc[1]+","+df_c.iloc[2]+","+df_c.iloc[3])
```

```python
# 색(color) 중 빨간색(red)이 가장 많은 공 크기(size)를 알아내는 코드를 작성하여 결과를 출력하라.
# 빨간색 공 중에서 어떤 크기의 공이 제일 많이 나왔는가?
#df[df['color']=='red' & df['size'].mode]
df_r = df[df['color'] =="red"] 
df_r['size'].mode()
```

```python
# df에서 색(color)과 크기(size)별 무게(weight)의 평균을 알아보는
# 피봇 테이블(df_pivot)을 만들고 출력하라

df_pivot = pd.pivot_table(df, index='color' ,columns='size', values='weight', aggfunc='mean')
df_pivot
```

```python
# df에서 공번호(number)에 대해 크기(size)별 각 색(color)의 갯수을 알아보는
# 피봇 테이블(df_pivot)을 만들고 출력하라

df_pivot = pd.pivot_table(df, index=['number','size'] ,columns='color', values='color', aggfunc='count',fill_value=0)
df_pivot
```

```python
# 상관관계 분석해보기
# 데이터 프레임(df)에서 공번호(number), 크기(size), 색(color), 무게(weight)의 관계를 알아보아라
# 정답은 없으며 어떻게 분석했는 지 과정이 중요함
```

```python
# df의 데이터를 공의 크기(size)별로 분류해
# 각각 df_small, df_medium, df_large로 분리하라

df_sm = df[df['size']=='small']
df_md = df[df['size']=='medium']
df_lg = df[df['size']=='large']

df_sm
```
